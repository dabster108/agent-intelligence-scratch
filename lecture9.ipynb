```python

from collections import deque
def bfs(graph, start_node, goal_node):
    queue = deque([start_node])
    parents = {start_node: None}
    visited_order = []
    step_trace = []

    while queue:
        current = queue.popleft()
        visited_order.append(current)

        discovered_now = []
        if current != goal_node:
            for neighbor in graph[current]:
                if neighbor not in parents:
                    parents[neighbor] = current
                    queue.append(neighbor)
                    discovered_now.append(neighbor)

        step_trace.append({
            "current": current,
            "visited": list(visited_order),
            "queue": list(queue),
            "discovered": discovered_now
        })

        if current == goal_node:
            break

    if goal_node not in parents:
        return step_trace, None

    path = []
    current = goal_node
    while current is not None:
        path.append(current)
        current = parents[current]
    path.reverse()
    return step_trace, path



tr, p = bfs(adj_list, start_node, goal_node)
tr
```

```python
def dfs(graph, start_node, goal_node):
    stack = [start_node]
    parents = {start_node: None}
    visited = set()
    
    visited_order = []
    step_trace = []

    while stack:
        current = stack.pop()

        if current in visited:
            continue

        visited.add(current)
        visited_order.append(current)

        discovered_now = []

        if current == goal_node:
            step_trace.append({
                "current": current,
                "visited": list(visited_order),
                "stack": list(stack),
                "discovered": discovered_now
            })
            break

        for neighbor in graph[current]:
            if neighbor not in visited and neighbor not in parents:
                parents[neighbor] = current
                stack.append(neighbor)
                discovered_now.append(neighbor)

        step_trace.append({
            "current": current,
            "visited": list(visited_order),
            "stack": list(stack),
            "discovered": discovered_now
        })

    # reconstruct path
    if goal_node not in parents:
        return step_trace, None

    path = []
    cur = goal_node
    while cur is not None:
        path.append(cur)
        cur = parents[cur]

    path.reverse()
    return step_trace, path
```

```python
import math

def heuristic(n1, n2):
    x1, y1 = G.nodes[n1]['x'], G.nodes[n1]['y']
    x2, y2 = G.nodes[n2]['x'], G.nodes[n2]['y']

    # simple Euclidean approximation (fast)
    return math.sqrt((x1 - x2)**2 + (y1 - y2)**2)
```

```python
import heapq

def a_star(graph, start_node, goal_node):
    open_set = []
    heapq.heappush(open_set, (0, start_node))

    parents = {start_node: None}
    g_cost = {start_node: 0}

    visited_order = []
    step_trace = []

    while open_set:
        _, current = heapq.heappop(open_set)

        visited_order.append(current)

        if current == goal_node:
            break

        discovered_now = []

        for neighbor in graph[current]:
            edge_weight = 1  # or use real distance if available

            tentative_g = g_cost[current] + edge_weight

            if neighbor not in g_cost or tentative_g < g_cost[neighbor]:
                g_cost[neighbor] = tentative_g
                f_cost = tentative_g + heuristic(neighbor, goal_node)

                heapq.heappush(open_set, (f_cost, neighbor))
                parents[neighbor] = current
                discovered_now.append(neighbor)

        step_trace.append({
            "current": current,
            "visited": list(visited_order),
            "open_set": [n[1] for n in open_set],
            "discovered": discovered_now
        })

    if goal_node not in parents:
        return step_trace, None

    path = []
    cur = goal_node
    while cur is not None:
        path.append(cur)
        cur = parents[cur]

    path.reverse()
    return step_trace, path
```

```python
astar_trace, astar_path = a_star(adj_list, start_node, goal_node)

print("A* Path Length:", len(astar_path) if astar_path else "Not Found")
print("A* Path:", astar_path)
```
